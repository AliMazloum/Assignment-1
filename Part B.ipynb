```python
plt.figure(figsize=(8, 6))

x = np.linspace(0,3,100) 
x0 = 3/2
y1 = 12/8 -(3/8)*x
y2 = 2 - x

# Plot the line at x1=3/2
plt.axvline(x0, label='x1=3/2')

# Plot line x1+(8/3)x2=4
plt.plot(x, y1, label='x1+(8/3)x2=4',color="orange")

# Plot line x1+x2=2
plt.plot(x, y2, label='x1+x2=2',color="green")


plt.scatter(1.5, 0.5,color="red",zorder=10)
plt.annotate('Minimizer\n(1.5, 0.5)', xy = (1.55, 0.5))

# Define the boundaries of the feasible area in the plot
y3 = np.minimum(y1, y2)
plt.fill_between(x,y3, where=x<1.5, color='grey', alpha=0.5)
plt.text(0.3, 0.5, 'Feasible Region', size = '11')

# Customize the plot
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Exercise 3')
plt.legend(fontsize=12)

# Set x-axis and y-axis values and limits
plt.xticks(np.arange(0, 3.1, 0.5),fontsize=12)  # Set x-axis values
plt.yticks(np.arange(0, 3.1, 0.5),fontsize=12)  # Set y-axis values
plt.xlim(0, 3)  # Set x-axis limits
plt.ylim(0, 3)  # Set y-axis limits
# plt.savefig('Exercise3.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def func(x):
    return 3 * x**4 - 4 * x**3 + 1

# Generate x values
x_values = np.linspace(-1, 1, 10000)

# Calculate corresponding y values
y_values = func(x_values)
y_values = np.array(y_values)

# Plot the function
plt.plot(x_values, y_values, label=r'$y = 3x^4 - 4x^3 + 1$')
# Point of inflection
plt.scatter(0.03535353535353536, 0.9998279369024238, color='red', marker='o')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Inflection point of the function $y = 3x^4 - 4x^3 + 1$')
plt.ylim(0, 3)  # Set y-axis limits
plt.xticks(np.arange(-1, 1.1, 0.2),fontsize=12)  # Set x-axis values
plt.yticks(np.arange(0, 3.1, 0.5),fontsize=12)  # Set y-axis values
# plt.legend()

# Display and save the plot
plt.savefig('Inflection_point.pdf')
# plt.show()
```
